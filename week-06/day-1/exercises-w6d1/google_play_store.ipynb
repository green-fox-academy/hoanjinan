```python
# What are the ten most recently updated apps?
df.sort_values(by = ["Last Updated"], ascending = False).iloc[:10]
```

```python
# Find apps, that is in the top 25% of apps based on ratings and were updated in the last two months of dataset.
from datetime import timedelta
twenty_five_percent = int(df.shape[0] * 0.25)
top_rating = df.sort_values(by = ["Rating"], ascending = False).iloc[:twenty_five_percent]
sorted_date = top_rating.sort_values(by = ["Last Updated"], ascending = False)
last_two_months = sorted_date["Last Updated"].max() - timedelta(days = 60)
sorted_date[sorted_date["Last Updated"] > last_two_months]
```

```python
# Which app not to download? - Find an app which takes the most space on your phone, but is rated poorly (define poorly)?
storage = []
for i in df["Size"]:
    if i.endswith("M"):
       storage.append(i)
storage.sort()
print(storage)
# df[df["Size"] != "Varies with device"].sort_values(by = ["Size"], ascending = False)
```
