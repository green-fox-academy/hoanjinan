```python
import requests
import json
import pandas as pd
from bs4 import BeautifulSoup
from pandas.io.json import json_normalize
```

```python
def get_souped(weblink):
    page = requests.get(weblink)
    soup = BeautifulSoup(page.content, "html.parser")
    return soup

root = "https://www.alza.co.uk"
soup = get_souped(root)
print(soup.prettify())
```

```python
# find_session = soup.findAll("div", {"id": "left"})
def find_link(name):
    link = soup.find("a", {"title": "Laptops"}).find_next_sibling().findAll("a", text = name)
    return link
home_n_office = find_link("Home & Office")[0]["href"]
gaming = find_link("Gaming")[0]["href"]
professional = find_link("Professional")[0]["href"]
# soup.body
```

```python
print(home_n_office)
print(gaming)
print(professional)
```

```python
def full_link(root, part_link):
    link = root + part_link
    return link

home_n_office = full_link(root, home_n_office)
gaming = full_link(root, gaming)
professional = full_link(root, professional)
```

```python
print(home_n_office)
print(gaming)
print(professional)
```

```python
home_soup = get_souped(home_n_office)
gaming_soup = get_souped(gaming)
professional_soup = get_souped(professional)
print(home_soup.prettify())
```

```python
def item_n_price(soup, index):
    name = soup.findAll("a", {"class": "name browsinglink"})[index].contents[0]
    old_price = soup.findAll("div", {"class": "price"})[index].find("div", {"class": "npc"}).find("span", {"class": "np2"})
    price = soup.findAll("div", {"class": "price"})[index].find("span", {"class": "c2"}).contents[0]
    if old_price == None:
        old_price = price
    else:
        old_price = old_price.contents[0]
    return name, old_price, price

def loop_by_soup(soup, soup_name):
    for i in range(3):
        name, old_price, price = item_n_price(soup, i)
        print(f"{soup_name}#{i + 1}\nProduct Name: {name}\nOriginal Price: {old_price}\nNew Price: {price}\n")

loop_by_soup(home_soup, "Home and Office")
loop_by_soup(gaming_soup, "Gaming")
loop_by_soup(professional_soup, "Professional")
# print(name)
# print(old_price)
# print(price)
```

```python
amazon = "https://www.amazon.co.uk/laptops/b?ie=UTF8&node=429886031&ref_=nav_shopall_lap"
amazon_soup = get_souped(amazon)
# print(soup_business.prettify())
```

```python
def amazon_item(index):
    name = amazon_soup.find("div", {"class":"a-section octopus-pc-card octopus-best-seller-card"}).findAll("span", {"class":"a-size-base a-color-base"})[index].contents[0].strip()
    
    symbol = amazon_soup.findAll("div", {"class": "a-section a-spacing-none octopus-pc-asin-price-section"})[index].find("span", {"class":"a-price-symbol"}).contents[0]
    whole = amazon_soup.findAll("div", {"class": "a-section a-spacing-none octopus-pc-asin-price-section"})[index].find("span", {"class":"a-price-whole"}).contents[0]
    decimal = amazon_soup.findAll("div", {"class": "a-section a-spacing-none octopus-pc-asin-price-section"})[index].find("span", {"class":"a-price-decimal"}).contents[0]
    fraction = amazon_soup.findAll("div", {"class": "a-section a-spacing-none octopus-pc-asin-price-section"})[index].find("span", {"class":"a-price-fraction"}).contents[0]
    
    amazon_price = symbol + whole + decimal + fraction

    check = amazon_soup.findAll("div", {"class": "a-section a-spacing-none octopus-pc-asin-price-section"})[index].find("span", {"class":"a-size-mini a-color-tertiary a-text-strike"})
    if check == None:
        amazon_old_price = amazon_price
    else:
        amazon_old_price = check.contents[0].strip()
    return name, amazon_old_price, amazon_price

for i in range(3):    
    item_name, amazon_old_price, amazon_price = amazon_item(i)
    print(f"#{i + 1}\nProduct Name: {item_name}\nOriginal Price: {amazon_old_price}\nNew Price: {amazon_price}\n")


```
